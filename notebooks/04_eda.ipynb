# 04: Exploratory Data Analysis (EDA) & Visualizations

We explore operational bottlenecks across airports, airlines, routes, and weather conditions using Matplotlib and Seaborn.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import sys
sys.path.append('..')
from src.config import PROCESSED_DATA_DIR
from src.utils.helpers import load_dataframe

sns.set_theme(style='whitegrid')
df = load_dataframe(PROCESSED_DATA_DIR / 'flights_features.parquet')
```

```python
# On-Time Performance by Airline
plt.figure(figsize=(10, 5))
ontime = df.groupby('airline_code')['arr_delay_mins'].apply(lambda x: (x < 15).mean() * 100).reset_index()
sns.barplot(x='arr_delay_mins', y='airline_code', hue='airline_code', data=ontime, palette='Greens_r', legend=False)
plt.title('On-Time Performance Rate (%) by Airline')
plt.show()
```

```python
# Delay breakdown across hours
hourly = df.groupby('scheduled_dep_hour')['arr_delay_mins'].mean().reset_index()
plt.figure(figsize=(10, 4))
plt.plot(hourly['scheduled_dep_hour'], hourly['arr_delay_mins'], marker='o', color='crimson', lw=2)
plt.title('Average Arrival Delay (Minutes) by Scheduled Departure Hour')
plt.xlabel('Scheduled Departure Hour')
plt.ylabel('Avg Arrival Delay (Mins)')
plt.show()
```
